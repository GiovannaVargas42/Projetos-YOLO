```python
import yaml
import os
dataset_path = "/content/drive/MyDrive/Colab Notebooks/Peixe/Fish Fry Dataset"
classes = ['alive', 'dead']
data_dict = {
    'path': dataset_path,
    'train': 'train/images/',
    'validation': 'validation/images/',
    'test': 'test/images/',  # Opcional
    'nc': len(classes),
    'names': classes
}
dataset_yaml_path = '/content/dataset.yaml'
with open(dataset_yaml_path, 'w') as f:
    yaml.dump(data_dict, f)
```

```python
with open(dataset_yaml_path, 'r') as f:
    print(f.read())
```

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8s.pt')

model.train(
    data=dataset_yaml_path,
    epochs= 40,
    imgsz= 640,
    batch= 16,
    name='fish_detection_yolov8',
    project='/content/drive/MyDrive/yolov8_runs'
)
```

```python
trained_model = YOLO('/content/drive/MyDrive/yolov8_runs/fish_detection_yolov82/weights/best.pt')
results = trained_model.val()

print(results)
```

```python
test_images = '/content/drive/MyDrive/Colab Notebooks/Peixe/Fish Fry Dataset/test/images/'

results_dir = '/content/drive/MyDrive/yolov8_runs/fish_detection_yolov8/inference/'

trained_model.predict(
    source= test_images,
    save=True,
    save_txt=True,
    conf=0.25,
    project= results_dir,
    name='detections'
)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

model_result = YOLO('/content/drive/MyDrive/yolov8_runs/fish_detection_yolov83/weights/best.pt')

uploaded_image_path = "/content/drive/MyDrive/Colab Notebooks/Peixe/Fish Fry Dataset/test/images/20_white_high_1dead_23.jpg"
results = model_result(uploaded_image_path)

result_image = results[0].plot()
plt.imshow(result_image)
plt.axis('off')
plt.show()
```
